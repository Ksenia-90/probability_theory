-**Задача 2**

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

В колоде 52 карт, из них 4 туза и 48 - нетузы
* вероятность первой карты нетуз 48/52
* вероятность второй карты нетуз 47/51
* вероятность третьей карты нетуз 46/50
* вероятность четвертой карты нетуз 45/49
* вероятность что из 4 карт хоть один был туз 
* P = 1-48 * 47 * 46 * 45 / (52 * 51 * 50 * 49) = 0,2813 = 28,13%

**Задача 3**

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?




Всего пар соседних скамеек - 6. Каждая пара человек может сесть двумя способами. Все остальные 5 любыми, то есть 5! способами.

Рассадить всех можем 7! способами

m=6 * 2 * 5!
n=7!

Вероятность того, что два определенных человека будут сидеть рядом : P=(6 * 2 * 5!)/7!=2/7

**Задача 5**

Бросается игральная кость. Пусть событие A - появление чётного числа, событие B - появление числа больше трёх. Являются ли эти события независимыми?

Событие А включает 3 исхода: A = {2; 4; 6}
* P(A) = 3/6 = 1/2 = 0,5


Событие B включает 3 исхода: B = {4; 5; 6}

P(A) = 3/6 = 1/2 = 0,5

(A ∩ B) = {4; 6}
* (A ∩ B) = 2/6 = 1/3

P(A | B) = P(A ∩ B)/P(B) = (1/3) / (1/2) = 2/3

События не являются независимыми

P(Болен | +) = P(+ | Болен) * P (Болен) / P(+| Болен)*P(Болен) + P(+| здоров)*P(Здоров) 

P(болен|+) = (0,99 * 0.001)/((0,99 * 0,001)+(0,01 * 0,999)) = 0,090163 ≈ 9,02%
